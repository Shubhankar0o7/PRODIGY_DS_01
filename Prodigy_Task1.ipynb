Top Ten Countries of Total Population

```python
total_popu_sorted = df[df['Indicator Code']=='SP.POP.TOTL']
total_popu_sorted = total_popu_sorted.sort_values(by='1999', ascending=False)
total_top_ten_countries = total_popu_sorted.head(10)
total_top_ten_countries[['Country Code']]
```

BAR PLOT

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="1999", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

Bottom Ten Countries of Total Population

```python
total_popu_sorted = df[df['Indicator Code']=='SP.POP.TOTL']
total_popu_sorted = total_popu_sorted.sort_values(by='1999', ascending=True)
total_bottom_ten_countries = total_popu_sorted.head(10)
total_bottom_ten_countries[['Country Code']]
```

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="1999", y="Country Code", data=total_bottom_ten_countries, palette="coolwarm")
plt.title("Bottom Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
